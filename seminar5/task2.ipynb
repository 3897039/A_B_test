Сравниваем метрику конверсия в покупку. Размер выборки - 10000 элементов в каждой группе . Какой статистический критерий тут лучше всего подойдёт и почему?

Согласно схеме выбираем:
тип исследуемых данных - колличественные - данные подчиняются нормальному закону распределения - :
1) сравнение с заданным значением - Z-test
2) независимые группы - 2 - Т-критерий Стьюдента для независимых выборок
Если данные не подчиняются нормальному закону распределени, то:
3) 2 независимые группы - критерий Манна-Уитни

В данной задаче больше всего подходит Z-тест, так как мы знаем размер выборки и тест основан на нормальном распределении и дает нам сравнить конверсии с использованием стандартного отклонения и смещением средней выборки.
